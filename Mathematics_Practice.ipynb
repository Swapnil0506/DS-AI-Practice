```python
'''
Q1
In a network, data packets are sent through a shared channel. Suppose there are 10 devices attempting to send data. In any given time slot, each device has a 20% chance of attempting to send a packet, independently of other devices. A collision occurs if two or more devices attempt to send a packet in the same time slot. What is the probability that in a given time slot, there will be no collisions?
'''
```

```python

import math
n = 10
p = 0.2

# P(X = 0)
p0 = (1 - p) ** n

# P(X = 1)
p1 = math.comb(n, 1) * (p ** 1) * ((1 - p) ** (n - 1))

prob_no_collision = p0 + p1
print("Probability of no collisions in a given time slot:", round(prob_no_collision, 4))
```

```python
'''
Q3
A software project has two modules, A and B. The probability that module A has a bug is 0.15, and the probability that module B has a bug is 0.10. The probability that both modules have bugs is 0.03.

(a) What is the probability that module A has a bug, given that module B has a bug?

(b) What is the probability that module B has a bug, given that module A does not have a bug?
'''
```

```python
P_A = 0.15
P_B = 0.10
P_A_and_B = 0.03

# (a) P(A | B)
P_A_given_B = P_A_and_B / P_B

# (b) P(B | A^c)
P_A_complement = 1 - P_A
P_B_and_A_complement = P_B - P_A_and_B
P_B_given_A_complement = P_B_and_A_complement / P_A_complement

print("P(A | B) =", P_A_given_B)
print("P(B | A^c) =", P_B_given_A_complement)
```

```python
'''
Q4  
A robot uses a sensor to detect an obstacle. The sensor has a 95% chance of correctly detecting an obstacle when one is present (true positive rate). However, it also has a 2% chance of indicating an obstacle when there isn't one (false positive rate). In the robot's environment, obstacles are present 10% of the time.

What is the probability that an obstacle is truly present, given that the sensor indicates an obstacle?
'''
```

```python

P_O = 0.10              # Obstacle present
P_not_O = 1 - P_O

P_S_given_O = 0.95     # True positive rate
P_S_given_not_O = 0.02 # False positive rate

# Bayes' Theorem
numerator = P_S_given_O * P_O
denominator = numerator + (P_S_given_not_O * P_not_O)

P_O_given_S = numerator / denominator

print(P_O_given_S)
```

```python
'''
Q7
A batch of 500 microchips is produced. The probability that a single microchip is defective is 0.02. Let X be the number of defective microchips in a random sample of 20 microchips from this batch. Assume the sample is taken with replacement, or the batch is large enough that we can model this as independent trials.

(a) What is the probability that exactly 3 microchips in the sample are defective? (b) What is the probability that at least 2 microchips in the sample are defective?
'''
```

```python
import math

n = 20
p = 0.02

# (a) P(X = 3)
P_3 = math.comb(n, 3) * (p ** 3) * ((1 - p) ** (n - 3))

# (b) P(X >= 2)
P_0 = (1 - p) ** n
P_1 = math.comb(n, 1) * p * ((1 - p) ** (n - 1))
P_at_least_2 = 1 - P_0 - P_1

print("P(X = 3) =", P_3)
print("P(X >= 2) =", P_at_least_2)
```

```python
'''
Q10
The latency (delay) in a specific network segment is observed to be normally distributed with a mean of 150 ms and a standard deviation of 20 ms.

(a) What is the probability that a data packet experiences a latency greater than 180 ms?

(b) Due to new QoS (Quality of Service) requirements, 90% of packets must arrive within a certain latency threshold. What is this threshold (in ms)?

(c) If a network engineer defines "critical latency" as any latency exceeding 2.5 standard deviations above the mean, what is the probability that a packet experiences critical latency?
'''
```

```python
import scipy.stats as stats

mu = 150
sigma = 20

# (a) P(X > 180)
P_gt_180 = 1 - stats.norm.cdf(180, mu, sigma)

# (b) 90% threshold
L_90 = stats.norm.ppf(0.9, mu, sigma)

# (c) Critical latency (> 2.5 std)
critical_latency = mu + 2.5 * sigma
P_critical = 1 - stats.norm.cdf(critical_latency, mu, sigma)

print("P(X > 180) =", P_gt_180)
print("Latency threshold for 90% packets =", L_90)
print("P(critical latency) =", P_critical)
```

```python
'''
Q11
The lifespan of a certain electronic component, due to manufacturing variations, is normally distributed with a mean of 2500 hours and a standard deviation of 300 hours.

(a) A customer requires components that last at least 2000 hours. What percentage of manufactured components meet this requirement?

(b) The manufacturer wants to provide a warranty such that only 1% of components fail before the warranty period ends. What should be the warranty period (in hours)?

(c) If a batch of 10,000 components is produced, how many components are expected to have a lifespan between 2200 and 2800 hours?
'''
```

```python
import scipy.stats as stats

mu = 2500
sigma = 300
total_components = 10000

# (a) P(X >= 2000)
P_ge_2000 = 1 - stats.norm.cdf(2000, mu, sigma)

# (b) Warranty period for 1% early failures
W = stats.norm.ppf(0.01, mu, sigma)

# (c) Number of components between 2200 and 2800
P_between = stats.norm.cdf(2800, mu, sigma) - stats.norm.cdf(2200, mu, sigma)
num_components = P_between * total_components

print("Percentage >= 2000 hours:", P_ge_2000)
print("Warranty period (hours):", W)
print("Expected components between 2200 and 2800 hours:", num_components)
```

```python
'''
Q12
The probability density function (pdf) of the triangular distribution is characterized by three points as shown in the figure below.

triangular distribution

(a) min value of the distribution, (b) max value of the distribution, and (c) mode of the distribution

Note: 

Create a class TriangularDistribution with attributes min, max and mode. You need to implement the following methods:

pdf(x): returns the value of the pdf at point 
.
plot_pdf(): plots the pdf in the range 
cdf(x): returns the value of the cdf at point 
 ; which denotes 
plot_cdf(): plots the cdf in the range 
get_probability(x1, x2): returns the probability 
, where 
.
Gracefully handle the possible errors and also the edge conditions when either 
 or 
 in the distribution.

Test your implementation.
'''
```

```python
import matplotlib.pyplot as plt
import numpy as np

class TriangularDistribution:
    def __init__(self, min_val, max_val, mode_val):
        if min_val > max_val:
            raise ValueError("min cannot be greater than max")
        if not (min_val <= mode_val <= max_val):
            raise ValueError("mode must be between min and max (inclusive)")
        
        self.min = min_val
        self.max = max_val
        self.mode = mode_val

        # Handle degenerate cases
        if self.min == self.max:
            self.is_degenerate = True
        else:
            self.is_degenerate = False

    def pdf(self, x):
        if self.is_degenerate:
            # Degenerate case: all mass at single point
            return 1.0 if x == self.min else 0.0

        if x < self.min or x > self.max:
            return 0.0
        
        if self.mode == self.min:
            # Mode at min edge
            if x == self.min:
                return 2 / (self.max - self.min)
            elif x <= self.max:
                return 2 * (self.max - x) / ((self.max - self.min) * (self.max - self.min))
        
        elif self.mode == self.max:
            # Mode at max edge
            if x == self.max:
                return 2 / (self.max - self.min)
            elif x >= self.min:
                return 2 * (x - self.min) / ((self.max - self.min) * (self.max - self.min))
        
        else:
            # Regular case
            if x <= self.mode:
                return 2 * (x - self.min) / ((self.max - self.min) * (self.mode - self.min))
            else:
                return 2 * (self.max - x) / ((self.max - self.min) * (self.max - self.mode))

    def cdf(self, x):
        if self.is_degenerate:
            return 0.0 if x < self.min else 1.0
        
        if x <= self.min:
            return 0.0
        elif x >= self.max:
            return 1.0
        
        if self.mode == self.min:
            if x < self.min:
                return 0.0
            elif x <= self.max:
                return 1 - ((self.max - x) ** 2) / ((self.max - self.min) ** 2)
        
        elif self.mode == self.max:
            if x <= self.min:
                return 0.0
            elif x < self.max:
                return ((x - self.min) ** 2) / ((self.max - self.min) ** 2)
        
        else:
            if x <= self.mode:
                return ((x - self.min) ** 2) / ((self.max - self.min) * (self.mode - self.min))
            else:
                return 1 - ((self.max - x) ** 2) / ((self.max - self.min) * (self.max - self.mode))

    def plot_pdf(self, num_points=1000):
        x = np.linspace(self.min, self.max, num_points)
        y = np.array([self.pdf(xi) for xi in x])
        plt.figure(figsize=(8, 4))
        plt.plot(x, y, label="PDF", color='blue')
        plt.title("Triangular Distribution PDF")
        plt.xlabel("x")
        plt.ylabel("pdf(x)")
        plt.grid(True)
        plt.legend()
        plt.show()

    def plot_cdf(self, num_points=1000):
        x = np.linspace(self.min, self.max, num_points)
        y = np.array([self.cdf(xi) for xi in x])
        plt.figure(figsize=(8, 4))
        plt.plot(x, y, label="CDF", color='green')
        plt.title("Triangular Distribution CDF")
        plt.xlabel("x")
        plt.ylabel("cdf(x)")
        plt.grid(True)
        plt.legend()
        plt.show()

    def get_probability(self, x1, x2):
        if x1 > x2:
            x1, x2 = x2, x1
        # Probability clipped to [min, max]
        x1 = max(x1, self.min)
        x2 = min(x2, self.max)
        if x2 < self.min or x1 > self.max:
            return 0.0
        return self.cdf(x2) - self.cdf(x1)
```

```python
# Regular triangular distribution test
tri = TriangularDistribution(min_val=0, max_val=10, mode_val=6)

print("PDF at x=5:", tri.pdf(5))
print("CDF at x=5:", tri.cdf(5))
print("Probability between 3 and 7:", tri.get_probability(3,7))

tri.plot_pdf()
tri.plot_cdf()

# Edge case: mode = min
tri_mode_min = TriangularDistribution(min_val=0, max_val=10, mode_val=0)
print("\nMode = min edge case:")
print("PDF at x=0:", tri_mode_min.pdf(0))
print("CDF at x=5:", tri_mode_min.cdf(5))
print("Probability between 0 and 5:", tri_mode_min.get_probability(0, 5))
tri_mode_min.plot_pdf()
tri_mode_min.plot_cdf()

# Edge case: mode = max
tri_mode_max = TriangularDistribution(min_val=0, max_val=10, mode_val=10)
print("\nMode = max edge case:")
print("PDF at x=10:", tri_mode_max.pdf(10))
print("CDF at x=5:", tri_mode_max.cdf(5))
print("Probability between 0 and 5:", tri_mode_max.get_probability(0, 5))
tri_mode_max.plot_pdf()
tri_mode_max.plot_cdf()

# Degenerate case: min = max = mode
tri_degenerate = TriangularDistribution(min_val=5, max_val=5, mode_val=5)
print("\nDegenerate case:")
print("PDF at x=5:", tri_degenerate.pdf(5))
print("PDF at x=4:", tri_degenerate.pdf(4))
print("CDF at x=4:", tri_degenerate.cdf(4))
print("CDF at x=5:", tri_degenerate.cdf(5))
print("Probability between 4 and 6:", tri_degenerate.get_probability(4,6))
tri_degenerate.plot_pdf()
tri_degenerate.plot_cdf()
```

```python
'''
Q13
The rand5() function which returns a uniform random integer within the range [1, 5], is defined as following:

def rand5():
    import numpy as np
    return np.random.randint(1,6)
Write a function rand7() that will generate a uniform random integer within the range [1, 7] by only using rand5().

Note: You can call rand5() as many time as you want inside the function rand7(), but you can not use np.random.randint(1, 8) or any other built-in methods in python.
'''
```

```python
def rand5():
    import numpy as np
    return np.random.randint(1, 6)

def rand7():
    while True:
        num = (rand5() - 1) * 5 + rand5()  # Uniform from 1 to 25
        if num <= 21:
            return (num - 1) % 7 + 1    
```

```python
'''
Q14
A grayscale image can be represented as a matrix where the dimensions are (height, width). Let's consider a simplified scenario with two grayscale images, 
 and 
, both represented as 10×10 matrices where each element represents pixel intensity. Pixel intensity is an integer in the range [0-255].

You want to create a blended image 
 by combining 
 and 
 such that 
. This represents a weighted average blend. Note that each element in the resultant matrix 
 should also fall within the range [0, 255].

write the code of the function blend_images(image_A, image_B, weight_A, weight_B) that takes two matrices and two scalar weights and produces another matrix as an output by performing the above operation.
'''
```

```python
import numpy as np

def blend_images(image_A, image_B, weight_A, weight_B):
    # Convert inputs to numpy arrays if they aren't already
    image_A = np.array(image_A)
    image_B = np.array(image_B)

    # Check dimensions
    if image_A.shape != image_B.shape:
        raise ValueError("Input images must have the same dimensions")
    
    # Weighted blend
    blended = weight_A * image_A + weight_B * image_B
    
    # Clip values to [0, 255]
    blended = np.clip(blended, 0, 255)
    
    # Convert to integers (pixel intensities)
    return blended.astype(np.uint8)
```

```python
# Create example 10x10 images
image_A = np.random.randint(0, 256, (10, 10))
image_B = np.random.randint(0, 256, (10, 10))

# Blend with weights
weight_A = 0.7
weight_B = 0.3

blended_image = blend_images(image_A, image_B, weight_A, weight_B)
print(blended_image)
```
