```python
'''
Q1
A startup company with 20 employees recorded their annual salaries (in INR Lakhs): [6.5, 7.2, 8.0, 6.8, 9.5, 7.0, 15.0, 7.5, 8.2, 6.9, 7.8, 10.5, 7.1, 8.5, 6.7, 7.3, 7.9, 9.0, 8.8, 7.6]

(a) Calculate the mean, median, and mode of the salaries.

(b) Determine the range, variance, and standard deviation of the salaries.

(c) Calculate the Interquartile Range (IQR). What does the IQR tell you about the salary distribution?

(d) Construct a box plot for the salaries. Identify any potential outliers.
'''
```

```python
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt

# Salary data
salaries = [6.5, 7.2, 8.0, 6.8, 9.5, 7.0, 15.0, 7.5, 8.2, 6.9,
            7.8, 10.5, 7.1, 8.5, 6.7, 7.3, 7.9, 9.0, 8.8, 7.6]


# (a) Mean, Median, Mode
mean_salary = np.mean(salaries)
median_salary = np.median(salaries)

try:
    mode_salary = stats.mode(salaries)
except stats.StatisticsError:
    mode_salary = "No unique mode"

print("Mean Salary:", round(mean_salary, 2))
print("Median Salary:", round(median_salary, 2))
print("Mode Salary:", mode_salary)


# (b) Range, Variance, Std Dev
salary_range = max(salaries) - min(salaries)
variance_salary = np.var(salaries, ddof=1)  # Sample variance
std_dev_salary = np.std(salaries, ddof=1)   # Sample standard deviation

print("\nRange:", round(salary_range, 2))
print("Variance:", round(variance_salary, 2))
print("Standard Deviation:", round(std_dev_salary, 2))


# (c) Interquartile Range (IQR)
Q1 = np.percentile(salaries, 25)
Q3 = np.percentile(salaries, 75)
IQR = Q3 - Q1

print("\nQ1:", round(Q1, 2))
print("Q3:", round(Q3, 2))
print("IQR:", round(IQR, 2))

print("\nInterpretation of IQR:")
print("The IQR represents the spread of the middle 50% of salaries.")
print("A larger IQR indicates more variability among the central salaries.")


# (d) Box Plot and Outliers
plt.boxplot(salaries, vert=False)
plt.title("Box Plot of Employee Salaries (INR Lakhs)")
plt.xlabel("Salary (INR Lakhs)")
plt.show()

# Identifying Outliers using IQR method
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in salaries if x < lower_bound or x > upper_bound]

print("\nLower Bound:", round(lower_bound, 2))
print("Upper Bound:", round(upper_bound, 2))
print("Potential Outliers:", outliers)
```

```python
'''
Q2
A tech company conducted a customer satisfaction survey, rating their latest product on a scale of 1 to 10 (10 being highly satisfied). The scores from 30 randomly selected customers are: [7, 8, 9, 6, 7, 8, 10, 5, 7, 8, 9, 7, 6, 8, 7, 9, 10, 6, 7, 8, 5, 9, 8, 7, 6, 7, 8, 9, 7, 8]

(a) Create a frequency distribution table for these scores.

(b) Plot a histogram of the customer satisfaction scores. Describe the shape of the distribution (e.g., symmetric, skewed).

(c) Calculate the Pearson's (first and second) coefficients of skewness for this data. Interpret the values.

(d) Calculate the excess kurtosis for this data using sample kurtosis formula. What does the kurtosis value suggest about the tails of the distribution?
'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Customer satisfaction scores
scores = [7, 8, 9, 6, 7, 8, 10, 5, 7, 8, 9, 7, 6, 8, 7,
          9, 10, 6, 7, 8, 5, 9, 8, 7, 6, 7, 8, 9, 7, 8]


# (a) Frequency Distribution Table
frequency = pd.Series(scores).value_counts().sort_index()
freq_table = pd.DataFrame({
    "Score": frequency.index,
    "Frequency": frequency.values
})

print("Frequency Distribution Table:\n")
print(freq_table)


# (b) Histogram
plt.hist(scores, bins=range(5, 12), edgecolor='black', align='left')
plt.xticks(range(5, 11))
plt.title("Histogram of Customer Satisfaction Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()

# Shape Description
print("\nDistribution Shape:")
print("The distribution is slightly left-skewed (negatively skewed) if mean < median.")


# (c) Pearson's Coefficients of Skewness
mean = np.mean(scores)
median = np.median(scores)
std_dev = np.std(scores, ddof=1)
mode = stats.mode(scores, keepdims=True)[0][0]

# Pearson’s First Coefficient (Mode Skewness)
pearson_first = (mean - mode) / std_dev

# Pearson’s Second Coefficient (Median Skewness)
pearson_second = 3 * (mean - median) / std_dev

print("\nMean:", round(mean, 3))
print("Median:", median)
print("Mode:", mode)
print("Standard Deviation:", round(std_dev, 3))

print("\nPearson's First Coefficient of Skewness:", round(pearson_first, 3))
print("Pearson's Second Coefficient of Skewness:", round(pearson_second, 3))

print("\nInterpretation:")
print("If skewness is negative → left-skewed.")
print("If skewness is positive → right-skewed.")
print("If close to 0 → approximately symmetric.")


# (d) Excess Kurtosis (Sample Kurtosis)
excess_kurtosis = stats.kurtosis(scores, fisher=True, bias=False)

print("\nExcess Kurtosis:", round(excess_kurtosis, 3))

print("\nInterpretation:")
print("Kurtosis > 0 → Leptokurtic (heavy tails)")
print("Kurtosis < 0 → Platykurtic (light tails)")
print("Kurtosis ≈ 0 → Mesokurtic (normal-like)")
```

```python
'''
Q3
A marketing team wants to understand the relationship between their monthly advertising expenditure (in INR Lakhs) and corresponding sales revenue (in INR Lakhs) over the last 10 months.

Month	Advertising Spend (X)	Sales Revenue (Y)
1	5	52
2	7	65
3	6	60
4	8	70
5	5	55
6	9	75
7	7	68
8	10	80
9	6	58
10	8	72
(a) Calculate the covariance between advertising spend and sales revenue.

(b) Plot Sales Revenue vs Advertising spent.

(c) Calculate the Pearson correlation coefficient between advertising spend and sales revenue.

(d) Interpret the meaning of both the covariance and the correlation coefficient in the context of this data.
'''

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Advertising Spend (X) and Sales Revenue (Y)
X = np.array([5, 7, 6, 8, 5, 9, 7, 10, 6, 8])
Y = np.array([52, 65, 60, 70, 55, 75, 68, 80, 58, 72])


# (a) Covariance
covariance_matrix = np.cov(X, Y)   # Sample covariance (default ddof=1)
covariance = covariance_matrix[0, 1]

print("Covariance between Advertising Spend and Sales Revenue:", round(covariance, 3))


# (b) Scatter Plot
plt.scatter(X, Y, color='blue')
plt.title("Sales Revenue vs Advertising Spend")
plt.xlabel("Advertising Spend (INR Lakhs)")
plt.ylabel("Sales Revenue (INR Lakhs)")
plt.grid(True)
plt.show()


# (c) Pearson Correlation Coefficient
correlation_matrix = np.corrcoef(X, Y)
correlation = correlation_matrix[0, 1]

print("Pearson Correlation Coefficient:", round(correlation, 3))


# (d) Interpretation
print("\nInterpretation:")
print("1. Covariance indicates the direction of the linear relationship.")
print("   Positive covariance → as advertising increases, sales tend to increase.")
print("2. Correlation coefficient measures strength and direction (range -1 to +1).")
print("   Values close to +1 indicate strong positive linear relationship.")
```

```python
'''
Q4
A traffic department claims that the average daily commute time for residents in a certain city is 45 minutes. A researcher believes this claim is too low and collects data from a random sample of 100 residents. The sample mean commute time is 48 minutes with a standard deviation of 12 minutes.

(a) Formulate the null and alternative hypotheses for this scenario.

(b) Construct a 95% confidence interval for the true average daily commute time.

(c) Based on the confidence interval, what can you conclude about the traffic department's claim?
'''
```

```python
import numpy as np
from scipy import stats

# Given data
sample_mean = 48
sample_std = 12
n = 100
claimed_mean = 45
confidence_level = 0.95


# (a) Hypotheses
print("(a) Hypotheses:")
print("Null Hypothesis (H0): μ = 45 minutes")
print("Alternative Hypothesis (H1): μ > 45 minutes")


# (b) 95% Confidence Interval
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

standard_error = sample_std / np.sqrt(n)
margin_of_error = t_critical * standard_error

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("\n(b) 95% Confidence Interval:")
print("Lower Bound:", round(lower_bound, 2))
print("Upper Bound:", round(upper_bound, 2))


# (c) Conclusion
print("\n(c) Conclusion:")
if claimed_mean < lower_bound or claimed_mean > upper_bound:
    print("The claimed mean (45 minutes) is NOT within the confidence interval.")
    print("We reject the traffic department's claim at 5% significance level.")
else:
    print("The claimed mean (45 minutes) lies within the confidence interval.")
    print("There is insufficient evidence to reject the claim.")
```

```python
'''
Q5
A manufacturing company wants to estimate the average number of units produced per hour by a new machine. They take 36 random observations and find the sample mean to be 150 units with a sample standard deviation of 15 units.

(a) What is the point estimate for the true average number of units produced per hour?

(b) Construct a 99% confidence interval for the true average number of units produced per hour.

(c) If the company wants to be 99% confident that their estimate is within 3 units of the true mean, what sample size would they need (assuming the same standard deviation)?
'''
```

```python
import numpy as np
from scipy import stats
import math

# Given data
sample_mean = 150
sample_std = 15
n = 36
confidence_level_99 = 0.99


# (a) Point Estimate
print("(a) Point Estimate:")
print("Point estimate for population mean =", sample_mean)


# (b) 99% Confidence Interval
alpha = 1 - confidence_level_99
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

standard_error = sample_std / np.sqrt(n)
margin_of_error = t_critical * standard_error

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("\n(b) 99% Confidence Interval:")
print("Lower Bound:", round(lower_bound, 2))
print("Upper Bound:", round(upper_bound, 2))


# (c) Required Sample Size
# Desired margin of error
E = 3

# For planning sample size, use Z value (large sample approximation)
z_critical = stats.norm.ppf(1 - alpha/2)

required_n = (z_critical * sample_std / E) ** 2
required_n = math.ceil(required_n)  # round up

print("\n(c) Required Sample Size for ±3 units at 99% confidence:")
print("Required sample size =", required_n)
```

```python
'''
Q6
A website designer believes that a new website redesign will increase the average time users spend on the site. Before the redesign, the average time spent was 3.5 minutes. After the redesign, a random sample of 50 users showed an average time of 3.8 minutes with a standard deviation of 1.0 minute.

(a) Perform a one-sided hypothesis test at a 0.05 significance level to determine if the redesign significantly increased the average time spent.

(b) State your null and alternative hypotheses, calculate the test statistic, and determine the p-value.

(c) What is your conclusion regarding the website redesign's impact?
'''
```

```python
import numpy as np
from scipy import stats

# Given data
population_mean = 3.5   # Before redesign
sample_mean = 3.8
sample_std = 1.0
n = 50
alpha = 0.05


# (a) & (b) Hypotheses and Test

print("(a) & (b) Hypothesis Test (One-Sided, α = 0.05)\n")

print("Null Hypothesis (H0): μ = 3.5 minutes")
print("Alternative Hypothesis (H1): μ > 3.5 minutes\n")

# Standard Error
standard_error = sample_std / np.sqrt(n)

# Test statistic (t)
t_statistic = (sample_mean - population_mean) / standard_error

# Degrees of freedom
df = n - 1

# One-sided p-value
p_value = 1 - stats.t.cdf(t_statistic, df=df)

print("Test Statistic (t):", round(t_statistic, 4))
print("Degrees of Freedom:", df)
print("P-value:", round(p_value, 6))


# (c) Conclusion
print("\n(c) Conclusion:")
if p_value < alpha:
    print("Reject H0.")
    print("There is sufficient evidence that the redesign increased average time spent.")
else:
    print("Fail to reject H0.")
    print("There is insufficient evidence that the redesign increased average time spent.")
```

```python
'''
Q7
A company implemented a new training program to improve employee productivity. To assess its effectiveness, 15 employees' productivity scores were recorded before and after the training program. The scores are as follows:

Employee	Before Training	After Training
1	65	70
2	70	72
3	60	68
4	75	73
5	80	85
6	68	70
7	72	75
8	62	65
9	78	80
10	70	72
11	66	69
12	73	76
13	64	67
14	71	74
15	69	71
(a) State the null and alternative hypotheses to test if the training program significantly improved productivity.

(b) Conduct a paired samples t-test at a 0.05 significance level. Calculate the test statistic and the p-value.

(c) Based on your analysis, did the training program effectively improve employee productivity?
'''
```

```python
import numpy as np
from scipy import stats

# Productivity scores
before = np.array([65, 70, 60, 75, 80, 68, 72, 62, 78, 70,
                   66, 73, 64, 71, 69])

after = np.array([70, 72, 68, 73, 85, 70, 75, 65, 80, 72,
                  69, 76, 67, 74, 71])

alpha = 0.05


# (a) Hypotheses
print("(a) Hypotheses:")
print("H0: μ_d = 0  (No improvement)")
print("H1: μ_d > 0  (Productivity improved)\n")


# (b) Paired t-test
# Differences (After - Before)
differences = after - before

mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)
n = len(differences)

# Standard Error
standard_error = std_diff / np.sqrt(n)

# Test Statistic
t_statistic = mean_diff / standard_error
df = n - 1

# One-sided p-value
p_value = 1 - stats.t.cdf(t_statistic, df=df)

print("(b) Test Results:")
print("Mean Difference:", round(mean_diff, 3))
print("Standard Deviation of Differences:", round(std_diff, 3))
print("t-statistic:", round(t_statistic, 4))
print("Degrees of Freedom:", df)
print("p-value:", round(p_value, 6))


# (c) Conclusion
print("\n(c) Conclusion:")
if p_value < alpha:
    print("Reject H0.")
    print("The training program significantly improved productivity.")
else:
    print("Fail to reject H0.")
    print("There is insufficient evidence that the training improved productivity.")
```

```python
'''
Q8
A battery manufacturer claims that their new 'ProLife' battery lasts longer than their existing 'Standard' battery. A consumer watchdog organization tests 40 'ProLife' batteries and 50 'Standard' batteries. The results are:

ProLife: Sample mean life = 120 hours, Sample standard deviation = 10 hours

Standard: Sample mean life = 115 hours, Sample standard deviation = 12 hours

(a) State the null and alternative hypotheses for testing the manufacturer's claim.

(b) Perform a t-test for two independent means at a 0.02 significance level.

(c) What is your conclusion regarding the manufacturer's claim about the 'ProLife' battery?
'''
```

```python
import numpy as np
from scipy import stats

# Given data
mean1 = 120   # ProLife
std1 = 10
n1 = 40

mean2 = 115   # Standard
std2 = 12
n2 = 50

alpha = 0.02


# (a) Hypotheses
print("(a) Hypotheses:")
print("H0: μ1 ≤ μ2  (ProLife does not last longer)")
print("H1: μ1 > μ2  (ProLife lasts longer)\n")


# (b) Welch's t-test (unequal variances)


# Standard Error
se = np.sqrt((std1**2 / n1) + (std2**2 / n2))

# Test Statistic
t_statistic = (mean1 - mean2) / se

# Welch-Satterthwaite Degrees of Freedom
df = ((std1**2 / n1 + std2**2 / n2)**2) / \
     (((std1**2 / n1)**2) / (n1 - 1) + ((std2**2 / n2)**2) / (n2 - 1))

# One-sided p-value
p_value = 1 - stats.t.cdf(t_statistic, df)

print("(b) Test Results:")
print("t-statistic:", round(t_statistic, 4))
print("Degrees of Freedom:", round(df, 2))
print("p-value:", round(p_value, 6))


# (c) Conclusion
print("\n(c) Conclusion:")
if p_value < alpha:
    print("Reject H0.")
    print("There is sufficient evidence at the 0.02 level that ProLife lasts longer.")
else:
    print("Fail to reject H0.")
    print("There is insufficient evidence at the 0.02 level to support the claim.")
```

```python
'''
Q9
A popular restaurant chain claims that the average waiting time for a table on weekends is no more than 20 minutes. A local food critic believes it's longer. They record the waiting times (in minutes) for 30 randomly selected tables on various weekends:

[22, 25, 18, 30, 28, 20, 23, 26, 19, 32, 21, 24, 27, 29, 20, 25, 23, 17, 31, 26, 22, 28, 20, 24, 27, 29, 21, 23, 25, 26]

(a) Calculate the sample mean and sample standard deviation of the waiting times.

(b) Conduct a one-sided t-test at a 0.01 significance level to assess the food critic's belief.

(c) State your null and alternative hypotheses, calculate the test statistic, and determine the critical value.

(d) Based on your findings, what can you tell the food critic?
'''
```

```python
import numpy as np
from scipy import stats

# Waiting times data
waiting_times = np.array([
    22, 25, 18, 30, 28, 20, 23, 26, 19, 32,
    21, 24, 27, 29, 20, 25, 23, 17, 31, 26,
    22, 28, 20, 24, 27, 29, 21, 23, 25, 26
])

claimed_mean = 20
alpha = 0.01
n = len(waiting_times)


# (a) Sample Mean & Std Dev
sample_mean = np.mean(waiting_times)
sample_std = np.std(waiting_times, ddof=1)

print("(a) Sample Statistics:")
print("Sample Mean:", round(sample_mean, 3))
print("Sample Standard Deviation:", round(sample_std, 3))


# (b) & (c) One-Sided t-test
print("\n(b) & (c) Hypothesis Test (α = 0.01)")

print("H0: μ ≤ 20 minutes")
print("H1: μ > 20 minutes")

# Standard Error
standard_error = sample_std / np.sqrt(n)

# Test Statistic
t_statistic = (sample_mean - claimed_mean) / standard_error
df = n - 1

# Critical value (right-tailed)
critical_value = stats.t.ppf(1 - alpha, df)

# One-sided p-value
p_value = 1 - stats.t.cdf(t_statistic, df)

print("\nTest Statistic (t):", round(t_statistic, 4))
print("Degrees of Freedom:", df)
print("Critical Value:", round(critical_value, 4))
print("p-value:", round(p_value, 6))


# (d) Conclusion
print("\n(d) Conclusion:")
if t_statistic > critical_value:
    print("Reject H0.")
    print("There is strong evidence that the average waiting time exceeds 20 minutes.")
else:
    print("Fail to reject H0.")
    print("There is insufficient evidence at the 1% level to conclude the wait exceeds 20 minutes.")
```

```python
'''
Q10
A software company is launching a new application and offers three different subscription tiers (Basic, Premium, Elite), each with varying features. Before the full launch, they conduct a pilot study to see if there's an association between a customer's age group and their preferred subscription tier. They surveyed 300 randomly selected potential customers and recorded their age group and preferred tier:

Age Group	Basic	Premium	Elite	Total
18-29 years	45	60	15	120
30-49 years	30	50	20	100
50+ years	25	40	15	80
Total	100	150	50	300
(a) State the null and alternative hypotheses for the test of independence in this context.

(b) Calculate the expected frequencies for each cell in the table, assuming the null hypothesis is true. Show your calculations clearly.

(c) Calculate the chi-square test statistic.

(d) Using a significance level of α=0.05, determine the critical value for this test.

(e) What is your conclusion? Is there a significant association between a customer's age group and their preferred subscription tier? Justify your answer based on your calculated chi-square statistic and the critical value.
'''
```

```python
import numpy as np
from scipy.stats import chi2

# Observed frequency table
# Rows: Age groups
# Columns: Subscription tiers
observed = np.array([
    [45, 60, 15],   # 18-29
    [30, 50, 20],   # 30-49
    [25, 40, 15]    # 50+
])

alpha = 0.05


# (a) Hypotheses
print("(a) Hypotheses:")
print("H0: Age group and subscription tier preference are independent.")
print("H1: There is an association between age group and subscription tier preference.\n")


# (b) Expected Frequencies
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
grand_total = observed.sum()

expected = np.outer(row_totals, col_totals) / grand_total

print("(b) Expected Frequencies:")
print(np.round(expected, 2))


# (c) Chi-square Test Statistic
chi_square_stat = np.sum((observed - expected)**2 / expected)

print("\n(c) Chi-square Statistic:", round(chi_square_stat, 4))


# (d) Critical Value
df = (observed.shape[0] - 1) * (observed.shape[1] - 1)
critical_value = chi2.ppf(1 - alpha, df)

print("(d) Degrees of Freedom:", df)
print("Critical Value at α=0.05:", round(critical_value, 4))


# (e) Conclusion
print("\n(e) Conclusion:")
if chi_square_stat > critical_value:
    print("Reject H0.")
    print("There is a significant association between age group and subscription preference.")
else:
    print("Fail to reject H0.")
    print("There is no significant association between age group and subscription preference.")
```
